```python
!ls /content
import pandas as pd

df = pd.read_csv("/content/Starter-Notebook/data/raw/content_refresh_anonymized.csv")



down_pct = (df["trend_direction"] == "down").mean() * 100

# 2. Difference in median engagement between growing and declining pages
median_engagement = df.groupby("trend_direction")["engagement_rate"].median()

# 3. Pages with above-median impressions but below-median CTR
opportunity_pages = df[
    (df["impressions_90d"] > df["impressions_90d"].median()) &
    (df["ctr"] < df["ctr"].median())
]

print(f"1. Pages trending downward: {down_pct:.1f}%")

print(
    "2. Median engagement rate:\n"
    f"   Growing pages: {median_engagement.get('up', float('nan')):.1%}\n"
    f"   Declining pages: {median_engagement.get('down', float('nan')):.1%}"
)

print(
    f"3. High-impression, low-CTR opportunity pages: "
    f"{len(opportunity_pages):,}"
)
```
